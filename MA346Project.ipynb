```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
concussion = pd.read_csv(r"Concussion Injuries 2012-2014.csv")

concussion['Date'] = pd.to_datetime(concussion["Date"], dayfirst=True)


concussion["Play Time After Injury"] = concussion["Play Time After Injury"].replace("Did not return from injury", 0)
concussion["Play Time After Injury"].fillna(0)
concussion['Play Time After Injury'] = concussion['Play Time After Injury'].str.replace(r'\D+', '', regex=True).fillna(0)
concussion['Play Time After Injury'] = concussion['Play Time After Injury'].astype(int)

# r'\D+' is saying it will search for non digits and then
# replace them with nothing, this strips down so its just numbers
# regex is just regular expression

concussion["Average Playtime Before Injury"].fillna(0)
concussion['Average Playtime Before Injury'] = concussion['Average Playtime Before Injury'].str.replace(r'\D+', '', regex=True).fillna(0)
concussion['Average Playtime Before Injury'] = concussion['Average Playtime Before Injury'].astype(int)

concussion["Games Missed"].fillna(0, inplace = True)
concussion.head()
```

```python
concussion.loc[concussion["Pre-Season Injury?"] == "Yes"]
```

```python
yr1213 = concussion.loc[concussion["Season"] == "2012/2013"]
yr1314 = concussion.loc[concussion["Season"] == "2013/2014"]
yr1415 = concussion.loc[concussion["Season"] == "2014/2015"]

fig, ax = plt.subplots(figsize = (8,8))
concussion["Season"].value_counts().plot(kind="bar")
ax.set_xlabel("Season")
ax.set_ylabel("Concussions")
ax.set_title("Concussions Per Season")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(concussion["Week of Injury"], bins=18, edgecolor = "black", color = 'green')
ax.set_xticks(range(1, 20))
ax.set_xlabel("Week of Injury")
ax.set_ylabel("Count")
ax.set_title("Week of Injury")
plt.show()

fig, ax = plt.subplots()
ax.hist(yr1213["Week of Injury"], bins=18, edgecolor = "black", color = 'blue', label = "2012/2013", alpha = .3)
ax.hist(yr1314["Week of Injury"], bins=18, edgecolor = "black", color = 'red', label = "2013/2014", alpha = .5)
ax.hist(yr1415["Week of Injury"], bins=18, edgecolor = "black", color = 'green', label = "2014/2015")
ax.set_xticks(range(1, 20))
ax.set_xlabel("Week of Injury")
ax.set_ylabel("Count")
ax.set_title("Week of Injury Per Season")
plt.legend(loc = "upper right")
plt.show()
```

```python
duplicateRows = concussion[concussion.duplicated(['Player'])]
alldups = duplicateRows['Player'].values.tolist()
multiple = []
[multiple.append(x) for x in alldups if x not in multiple]
multiple
# This sets all duplicate names into a list to pull up all duplicates
# .duplicate only shows the duplicates not the first one as well
```

```python
multiconcussion = concussion.loc[concussion['Player'].isin(multiple)]
multiconcussion
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.barh(multiconcussion['Player'], multiconcussion["Games Missed"], edgecolor = "black", color = 'blue', alpha=.5)
ax.set_xlabel("Games Missed")
ax.set_ylabel("Count")
ax.set_title("Games missed When multiple concussions")
plt.show()
```
